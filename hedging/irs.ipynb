```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Define the data for the three instruments
instruments = [
    {'Instrument': 'Interest Rate Swaps', 'Potential Returns': 'High', 'Risk Exposure': 'High'},
    {'Instrument': 'Options on Interest Rate Futures', 'Potential Returns': 'High', 'Risk Exposure': 'Low'},
    {'Instrument': 'Inverse ETFs', 'Potential Returns': 'Low', 'Risk Exposure': 'High'}
]

# interest rate, options, inverse ETF
data = {
    'Exposure': [0.3, 0.5, 0.7], 
    'Returns': [0.6, 0.4, 0.3]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create a scatterplot with Seaborn
plt.figure(figsize=(8, 6))

# Set style
sns.set_theme(style="white")

# Create scatter plot
sns.scatterplot(data=df, x='Returns', y='Exposure', s=100)

# Add horizontal and vertical lines at the midpoint
plt.axhline(y=0.5, color='black', linestyle='-', linewidth=2)
plt.axvline(x=0.5, color='black', linestyle='-', linewidth=2)


# Set axis limits
plt.xlim(0, 1)
plt.ylim(0, 1)

# Add labels to quadrants
plt.text(0.4, 1.03, 'High Risk Exposure', fontsize=12, color='black')
plt.text(0.4, -0.04, 'Low Risk Exposure', fontsize=12, color='black')
plt.text(0.78, 0.52, 'High Potential Return', fontsize=12, color='black')
plt.text(-0.03, 0.52, 'Low Potential Return', fontsize=12, color='black')

# hide outer axes
sns.despine() 
plt.axis('off')
# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Define the size of the square
size = 10

# Create a gradient square with a diagonal angle
gradient = np.zeros((size, size))

# Populate the gradient array with values ranging from 0 to 1 along the diagonal
for i in range(size):
    for j in range(size):
        gradient[i, j] = (i + j) / (size - 1)

# Create a heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(gradient, cmap='RdYlGn', square=True, cbar=False)

# Set axis labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the size of the square
size = 100

# Create a smooth gradient square with a diagonal angle
x = np.linspace(0, 1, size)
y = np.linspace(0, 1, size)
X, Y = np.meshgrid(x, y)
gradient = X + Y

# Create a plot
plt.figure(figsize=(6, 6))
plt.imshow(gradient, cmap='RdYlGn', interpolation='spline16')

# Remove axis ticks
plt.xticks([])
plt.yticks([])

# Show the plot
plt.show()
```
