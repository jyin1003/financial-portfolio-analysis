```python
# Lexicon-based sentiment analysis on web-scraped news articles

from pygooglenews import GoogleNews
from newscatcher import Newscatcher

import datetime

gn = GoogleNews()
```

```python
# Use pygooglenews to find news articles 
top = gn.top_news()

# set time constraints
start_date = datetime.date(2022, 1, 1)
today = datetime.date.today()
start_date_str = start_date.strftime('%Y-%m-%d')
today_str = today.strftime('%Y-%m-%d')


# set search filters and add to list of articles to scrape
articles = []
filtered_news = gn.geo_headlines('("ukraine" OR "russia") \
    allintext:(invasion OR conflict OR war OR crisis)')
dated_news = gn.search('allintext:(invasion OR conflict OR war OR crisis + ukraine + russia)', from_ = start_date_str, to_ = today_str)
for entry in dated_news["entries"]:
    #print(f"title: {entry['title']}")
    #print(f"date published: {entry['published']}")
    #print(f"link: {entry['link']}")
    articles.append((entry["published"], entry["link"]))

sorted_date_link_tuples = sorted(articles, key=lambda x: x[0], reverse=True)
print(sorted_date_link_tuples)

```
