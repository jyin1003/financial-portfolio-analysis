```python
# Lexicon-based sentiment analysis on web-scraped news articles
from modules.pygooglenews import GoogleNews
import inspect
from html import unescape
import feedparser
from newscatcher import Newscatcher
import requests
from bs4 import BeautifulSoup
import base64
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

from newspaper import Article

from datetime import datetime, timedelta, date
from collections import Counter
from dateutil import parser
import re


def parse_date(date_str):
    pattern = r'\b(\d+)\s+(month|months|week|weeks|day|days|hour|hours)\s+ago\b'
    match = re.search(pattern, date_str)
    if match:
        current_date = datetime.now()
        quantity = int(match.group(1))
        unit = match.group(2)
        if unit == 'months' or unit == 'month':
            return current_date - timedelta(days=30 * quantity)
        elif unit == 'weeks' or unit == 'week':
            return current_date - timedelta(weeks=quantity)
        elif unit == 'days' or unit == 'day':
            return current_date - timedelta(days=quantity)
        elif unit == 'hours' or unit == 'hour':
            return current_date - timedelta(hours=quantity)
    else:
        return parser.parse(date_str)
        
    #return datetime.strptime(date_str, '%b %d, %Y')

gn = GoogleNews()
```
